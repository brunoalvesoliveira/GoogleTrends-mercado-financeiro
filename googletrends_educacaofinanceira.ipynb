```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
from matplotlib import style
```

```python
style.use('seaborn-muted')
```

Buscando os dados

```python
pytrends = TrendReq(hl='en_US', tz=360)
```

Palavras-chave

```python
lista = ['aposta', 'investimento']
```

Buscando pelas palavras-chave

```python
pytrends.build_payload(lista, cat=0, timeframe='2019-01-01 2021-10-07', geo='BR', gprop='')
```

```python
dados = pytrends.interest_over_time()
dados.head()
```

Plotando o gr√°fico

```python
plt.figure(figsize=(12, 6));
plt.title('Pesquisas no Google: Aposta versus Investimento', fontsize=15);
x, = plt.plot(dados.aposta, color='red');
y, = plt.plot(dados.investimento, color='green');
plt.xlabel('Data', fontsize=13);
plt.ylabel('Freq. de pesquisa', fontsize=13);
plt.text(np.datetime64('2020-12-01'), 15, 'By: @financeiroarretado - GoogleTrends', fontsize=10);
plt.legend([x, y], ['Aposta', 'Investimento'], fontsize=10, loc=2);
```
